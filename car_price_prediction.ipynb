```python
x=a.drop(["Car ID","Price"],axis=1)
x.head()
```

```python
#encode the character to categorical
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
for i in ["Brand","Fuel Type","Transmission","Condition","Model"]:
    a[i]=label.fit_transform(a[i])
a.head()
```

```python
y=a["Price"]
y.head()
```

```python
#train,test and split the data for x and y
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

#linear regression
ln = LinearRegression()
ln.fit(x_train, y_train)

ypred = ln.predict(x_test)
#print(ypred)

r2 = r2_score(y_test, ypred)
rmse = np.sqrt(mean_squared_error(y_test, ypred))  # calculate RMSE manually

print("Linear Regression Results:")
print("R2 Score:", r2)
print("RMSE:", rmse)
#Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
ypred_rf = rf.predict(x_test)

r2_rf = r2_score(y_test, ypred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, ypred_rf))

print("\nRandom Forest Regressor Results:")
print("R2 Score:", r2_rf)
print("RMSE:", rmse_rf)
```
